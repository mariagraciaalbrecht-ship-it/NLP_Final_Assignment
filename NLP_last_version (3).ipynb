```python
# Code obtained from Chat GPT
def coerce_missing(series: pd.Series) -> pd.Series:
    s = series.astype("string").str.normalize("NFKC").str.strip()
    # Mark as missing any placeholders
    mask = s.str.len().eq(0) | s.str.fullmatch(r'(?i)(n/?a|none|null|missing|unknown|-||)')
    s = s.mask(mask, pd.NA)
    return s

df["speaker_party"] = coerce_missing(df["speaker_party"])
```

# 1.LLM-based structured extraction (on a subset)

I will only keep the interventions carried out originally in Spanish from the years 2018 to 2022.

```python
# Keeping only speeches that were originally delivered in Spanish
# Code fixed/troubleshoot by Google Gemini
df_spanish = df[df["intervention_language"] == "es"].copy()

# The column "year" was originally an object. Extract the year from the 'date' column in df_spanish
df_spanish["year"] = df_spanish["date"].dt.year

# Keeping only data from 2018 to 2022
df_spanish = df_spanish[(df_spanish["year"] >= 2018) & (df_spanish["year"] <= 2022)].copy()

print("Filtered data:")
df_spanish.head(3)
```

First, I will use the LLM to extract the year, party and topic from the texts from the interventions carried out in the debates.

```python
# Code adapted from the m2_nlp_llmstructure.ipynb notebook shown in class, aided by Chat GPT

class PartyYearTopic(BaseModel):
    party: Optional[str] = Field(default=None, description="Party/group as stated in the speech text; null if not stated. Write it in English")
    year: Optional[int]  = Field(default=None, description="4-digit year if stated/implied in the text; else null.")
    topic: str           = Field(description="Main issue in 25 English words.")

pyt_schema = PartyYearTopic.model_json_schema()

triplet_prompt = """
Extract the speaker's party, the year, and the main issue discussed from the following text, in English.
Provide ONLY a JSON object with keys "party", "year", "topic".
If party or year are NOT stated or cannot be inferred from the speech TEXT itself, set them to null.
The "topic" must be a concise 25 word English label.

SCHEMA:
{schema}

Text: \"\"\"{text}\"\"\""
"""

def llm_extract_party_year_topic(text: str, model_name="gemini-2.5-flash-lite"):
    prompt = triplet_prompt.format(
        schema=json.dumps(pyt_schema, indent=2),
        text=(text[:5000] if isinstance(text, str) else "")
    )
    time.sleep(0.5)
    completion = client.chat.completions.create(
        model=model_name,
        messages=[{"role": "user", "content": prompt}],
        response_format={
            "type": "json_schema",
            "json_schema": {"name": "PartyYearTopic", "strict": True, "schema": pyt_schema}
        },
        temperature=0.0,
        timeout=60
    )
    raw = completion.choices[0].message.content
    return PartyYearTopic.model_validate_json(raw)

# Run on a sample
SAMPLE1 = min(100, len(df_spanish))
df_s1 = df_spanish.sample(SAMPLE1, random_state=1).reset_index(drop=True)

rows1 = []
for _, r in tqdm(df_s1.iterrows(), total=len(df_s1), desc="LLM-only: party, year, topic"):
    txt = str(r["text"]).strip()
    if not txt:
        continue
    try:
        out = llm_extract_party_year_topic(txt)
        d = out.model_dump()

        d["_ds_party"] = r["speaker_party"]
        d["_ds_year"]  = int(r["year"]) if pd.notna(r["year"]) else None
        rows1.append(d)
    except Exception:
        pass

llm_only = pd.DataFrame(rows1)
llm_only.head()
```

```python
# Saving it as a .csv file
# Code obtained from Chat GPT
llm_only.to_csv("llm_only_extraction.csv", index=False)

print(" Saved as: llm_only_extraction.csv")
```

Even though the table presented is only a small preview, I can see that the LLM has extracted the party correctly from some of the cases. However, it does not always extract the party name correctly and it almost never extracts the year. This is most likely because the speakers usually do not mention the current year or the name of their own party in their speeches.

Therefore, I will only extract the topic from the text using the LLM, and will keep the year and the party name from the database. This will be called the "hybrid" because it combines extraction (topic) and keeps other important information like the year and the party name the speaker belongs to.

```python
# Code adapted from the m2_nlp_llmstructure.ipynb notebook shown in class, aided by Chat GPT

class SpeechTopic(BaseModel):
    topic: str = Field(description="Main issue/topic in 25 English words.")

topic_schema = SpeechTopic.model_json_schema()

topic_prompt = """
Read the following European Parliament speech in Spanish and return ONLY valid JSON per the SCHEMA.
Rules: output a concise English topic (25 words). No commentary.

SCHEMA:
{schema}

Text: \"\"\"{text}\"\"\""
"""

def llm_extract_topic(text: str, model_name="gemini-2.5-flash-lite"):
    prompt = topic_prompt.format(
        schema=json.dumps(topic_schema, indent=2),
        text=(text[:5000] if isinstance(text, str) else "")
    )
    time.sleep(0.5)
    completion = client.chat.completions.create(
        model=model_name,
        messages=[{"role": "user", "content": prompt}],
        response_format={
            "type": "json_schema",
            "json_schema": {"name": "SpeechTopic", "strict": True, "schema": topic_schema}
        },
        temperature=0.0,
        timeout=60
    )
    raw = completion.choices[0].message.content
    return SpeechTopic.model_validate_json(raw)

SAMPLE2 = min(100, len(df_spanish))
df_s2 = df_spanish.sample(SAMPLE2, random_state=2).reset_index(drop=True)

rows2 = []
for _, r in tqdm(df_s2.iterrows(), total=len(df_s2), desc="Hybrid: topic only"):
    txt = str(r["text"]).strip()
    if not txt:
        continue
    try:
        out = llm_extract_topic(txt)
        rows2.append({
            "party": r["speaker_party"],
            "year": int(r["year"]) if pd.notna(r["year"]) else None,
            "topic": out.topic
        })
    except Exception:
        pass

hybrid = pd.DataFrame(rows2)
hybrid.head()
```

```python
# Saving it as a .csv file
# Code obtained from Chat GPT
hybrid.to_csv("hybrid_llm_extraction.csv", index=False)
print(" Saved as: hybrid_llm_extraction.csv")
```

```python
!pip -q install bertopic sentence-transformers umap-learn hdbscan nltk

# Code obtained from Chat GPT
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
spanish_stopwords = stopwords.words("spanish")

df_bt = df_spanish.copy()
df_bt["doc"] = df_bt["text"].astype(str).str.strip()
df_bt = df_bt[df_bt["doc"].str.len() > 20].reset_index(drop=True)

docs = df_bt["doc"].tolist()

# Embeddings
sbert_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

# Spanish vectorizer
spanish_vectorizer = CountVectorizer(
    stop_words=spanish_stopwords,
    ngram_range=(1, 2),
    min_df=10
)

# Build BERTopic model
topic_model = BERTopic(
    embedding_model=sbert_model,
    vectorizer_model=spanish_vectorizer,
    language="multilingual",
    calculate_probabilities=True,
    nr_topics="auto",
    verbose=False
)

topics, probs = topic_model.fit_transform(docs)
df_bt["topic_id"] = topics

topic_info = topic_model.get_topic_info()
topic_info.head(10)
```

```python
# Code obtained from Chat GPT

class TopicLabel(BaseModel):
    label: str = Field(description="Short 24 word English topic name.")
```

```python
# Code obtained from ChatGPT

def llm_label_topic(words, reps, model="gemini-2.5-flash-lite"):
    prompt = f"""
You are a political analyst.
Summarize the following topic into a short 24 word English label.

Top words: {words}
Representative texts: {reps[:2]}  # show first 2 examples only

Return ONLY valid JSON:
{TopicLabel.model_json_schema()}
    """

    out = client.chat.completions.create(
        model=model,
        response_format={
            "type": "json_schema",
            "json_schema": {"name": "TopicLabel", "strict": True,
                            "schema": TopicLabel.model_json_schema()}
        },
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2
    )

    return TopicLabel.model_validate_json(out.choices[0].message.content).label
```

```python
# Code obtained from ChatGPT

final_labels = {}

for topic_id in topic_info["Topic"]:
    if topic_id == -1:
        continue

    topic_data = topic_model.get_topic(topic_id)
    words = [w for w, _ in topic_data]
    reps = topic_model.get_representative_docs(topic_id)

    final_labels[topic_id] = llm_label_topic(words, reps)
```

## Main limitations

Regarding the limitations and systematic errors I have observed, the LLM alone can produce a lot of labels, and can produce a unique topic label for each specific text. It can make the label very accurate and very precise, which makes it able to encompass more than one topic per text, which debates and speeches usually do.

When it comes to using BERTopic and then an LLM to help summarize the labels, the topics can be too broad. BERTopic finds similarities and can identify main themes, and it can understand context so it is useful for finding similarities between synonyms or concepts that are related. The LLM will work on top of what BERTopic does, and in this case, the topics identified ended up being too broad and missing the more important themes that the speeches in the debate were dealing with.

# 3. Building a knowledge graph

To build the knowledge graph, I will use the following elements:


*   Nodes: Party and topics
*   Edges: Party and topic --> The relationship is: The party mentions the topic
*   Edge attribute: Time (year)







```python
# Code adapted from the 1_KG_clean.ipynb notebook shown in class, aided by Chat GPT

df_kg = df_bt_clean.copy()
df_kg = df_kg.dropna(subset=["speaker_party", "year", "topic_label"]).copy()

PartyTopicsKG = nx.Graph()

# Add topic nodes
for topic in df_kg["topic_label"].unique():
    topic_id = f"t_{topic}"
    PartyTopicsKG.add_node(
        topic_id,
        type="Topic",
        label=topic
    )

# 2. Add party nodes and "mentions_topic" edge
for _, row in df_kg.iterrows():
    party_id = f"p_{row["speaker_party"]}"
    topic_id = f"t_{row["topic_label"]}"

    # Add the party node
    PartyTopicsKG.add_node(party_id, type="Party", name=row["speaker_party"])

    # Add the edge connecting the party and the topic
    PartyTopicsKG.add_edge(
        party_id,
        topic_id,
        relation="mentions_topic",
        year=int(row["year"])
    )

print(f" Knowledge Graph built successfully!")
print(f"Total nodes: {PartyTopicsKG.number_of_nodes():,}")
print(f"Total edges: {PartyTopicsKG.number_of_edges():,}")
```

```python
# Code adapted from the 1_KG_clean.ipynb notebook shown in class, aided by Chat GPT
def draw_party_topic_subgraph(graph, party_id):
    """
    Creates and draws a subgraph for a specific party, showing
    all topics that the party is connected to.
    """
    if not graph.has_node(party_id):
        print(f"Error: Party '{party_id}' not found.")
        return

    # Get all topics connected to the party
    topic_nodes = list(graph.neighbors(party_id))
    subgraph_nodes = [party_id] + topic_nodes

    # Create the subgraph containing only these nodes
    subG = graph.subgraph(subgraph_nodes).copy()

    # Prepare layout weights
    for u, v, data in subG.edges(data=True):

        data["layout_weight"] = 1.0

    # Setting up the visualization
    plt.figure(figsize=(16, 16))

    # Use the layout_weight as the spring layout weight
    pos = nx.spring_layout(subG, seed=42, k=0.8, weight="layout_weight", iterations=50)

    # Color nodes by type (Party or Topic)
    node_colors = [
        "red" if subG.nodes[n].get("type") == "Party" else "skyblue"
        for n in subG.nodes()
    ]

    # Node labels:
    # Party -> its 'name'; Topic -> its 'label'
    node_labels = {}
    for n in subG.nodes():
        n_type = subG.nodes[n].get("type")
        if n_type == "Party":
            node_labels[n] = subG.nodes[n].get("name", n)
        elif n_type == "Topic":
            node_labels[n] = subG.nodes[n].get("label", n)
        else:
            node_labels[n] = n

    # Draw nodes and labels
    nx.draw_networkx_nodes(subG, pos, node_color=node_colors, node_size=3000)
    nx.draw_networkx_labels(subG, pos, labels=node_labels, font_size=10, font_weight="bold")

    # Edge widths (1)
    edge_widths = [1.5 for _, _, _ in subG.edges(data=True)]
    nx.draw_networkx_edges(subG, pos, width=edge_widths, alpha=0.7, edge_color="gray")

    # Edge labels, show year if present
    edge_labels = nx.get_edge_attributes(subG, "year")
    nx.draw_networkx_edge_labels(subG, pos, edge_labels=edge_labels,
                                 font_color="black", font_size=9)

    plt.title(f"Topic Universe for {party_id}", size=18)
    plt.axis("off")
    plt.show()
```

## 4.2 Comparing networks before and after COVID19

I also want to compare the most important topics before and after COVID19. For this, I thought of splitting the dataset into 2 dataframes: before and after the pandemic. Then, I will use degree centrality for both of them.

```python
# Topic degree centrality from before and after COVID19
# Code obtained from Chat GPT

# Splitting the dataset
df_net = df_bt_clean.copy()
df_net = df_net.dropna(subset=["speaker_party", "topic_label", "year"]).copy()
df_before = df_net[(df_net["year"] >= 2018) & (df_net["year"] <= 2019)].copy()
df_after  = df_net[(df_net["year"] >= 2020) & (df_net["year"] <= 2022)].copy()


# Build the two networks
G_before = build_party_topic_network(df_before)
G_after  = build_party_topic_network(df_after)

# --- Degree for PRE-COVID network (20182019) ---
cent_degree_before = dict(nx.degree(G_before))

# Store as node attribute
nx.set_node_attributes(G_before, cent_degree_before, 'cent_degree')

# Optional: inspect
nx.get_node_attributes(G_before, 'cent_degree')

# Filter to Topic nodes only
topics_before = [
    (n, d) for n, d in cent_degree_before.items()
    if G_before.nodes[n].get("type") == "Topic"
]

# Top 10 topics by degree
top_topics_before = sorted(topics_before, key=lambda x: x[1], reverse=True)[:10]

top_topics_before_df = pd.DataFrame(
    {
        "node_id": [n for n, _ in top_topics_before],
        "degree": [d for _, d in top_topics_before],
        "topic_label": [G_before.nodes[n].get("label") for n, _ in top_topics_before],
    }
)
print(" Top topics PRE-COVID (20182019):")
display(top_topics_before_df)


# --- Degree for POST-COVID network (20202022) ---
cent_degree_after = dict(nx.degree(G_after))

nx.set_node_attributes(G_after, cent_degree_after, 'cent_degree')
nx.get_node_attributes(G_after, 'cent_degree')

topics_after = [
    (n, d) for n, d in cent_degree_after.items()
    if G_after.nodes[n].get("type") == "Topic"
]

top_topics_after = sorted(topics_after, key=lambda x: x[1], reverse=True)[:10]

top_topics_after_df = pd.DataFrame(
    {
        "node_id": [n for n, _ in top_topics_after],
        "degree": [d for _, d in top_topics_after],
        "topic_label": [G_after.nodes[n].get("label") for n, _ in top_topics_after],
    }
)
print(" Top topics POST-COVID (20202022):")
display(top_topics_after_df)
```

However, I do not fully think this is the case, and I think this might be mostly because of the limitations from BERTopic and the LLM.
